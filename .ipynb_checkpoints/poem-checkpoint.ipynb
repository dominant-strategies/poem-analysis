## Honest Party execution

```python
import csv
import pandas as pd
```

```python
file_honest='/home/gop/quai/POEM-analysis/poemhonest/2023112829/34.121.121.229/simulations/1.csv'
```

```python
dfh = pd.read_csv(file_honest, skiprows=1)
```

```python
dfh
```

```python
dfh['Delta Timestamp']=dfh[' Time Stamp'].diff()/1000
dfh['Delta Timestamp']=dfh['Delta Timestamp'].fillna(0)
```

```python
dfh['CumBits']=dfh[' Extra Bits'].cumsum()
dfh['CumTime']=dfh['Delta Timestamp'].cumsum()
```

```python
dfh
```

```python
rows, columns = dfh.shape
```

```python
import math
x_bits=[]
for i in range(0, rows):
    x_bits.append(i/math.log(2))
```

```python
def filter_first_le_value(df, low):
    filtered_df = df[df['CumBits'] > low]
    if not filtered_df.empty:
        return filtered_df.iloc[0]
    else:
        return None
```

```python
x_bits[1], x_bits[2], x_bits[3]
```

```python
filter_first_le_value(dfh, x_bits[3])
```

```python
cum_times=[]
cum_bits=[]
```

```python
for bits in x_bits:
    value=filter_first_le_value(dfh, bits)
    cum_times.append(value.CumTime)
    cum_bits.append(value.CumBits)
```

```python
len(x_bits), len(cum_times), len(cum_bits)
```

```python
columns=['Bits', 'Time', 'CumBits']
dfh_final = pd.DataFrame(columns=columns)

dfh_final['Bits']=x_bits
dfh_final['Time']=cum_times
dfh_final['CumBits']=cum_bits
dfh_final
```

## Monte Carlo Analysis and Probability Calculations

```python
folder_honest="/home/gop/quai/POEM-analysis/poemhonest/2023112829/34.121.121.229/simulations/"
folder_adv="/home/gop/quai/POEM-analysis/poemadv/2023112839/34.71.126.0/simulations/"
```

```python
import math
def compute_one_round(i):
    # Honest Party
    honest_file=folder_honest+f"{i}"+".csv"
    df1 = pd.read_csv(honest_file, skiprows=1)
    df1['Delta Timestamp']=df1[' Time Stamp'].diff()/1000
    df1['Delta Timestamp']=df1['Delta Timestamp'].fillna(0)
    df1['CumBits']=df1[' Extra Bits'].cumsum()
    df1['CumTime']=df1['Delta Timestamp'].cumsum()


    rows, columns = df1.shape
    x_bits=[]
    for i in range(0, rows):
        x_bits.append(i/math.log(2))

    columns=['Bits', 'Time', 'CumBits']
    df1_final = pd.DataFrame(columns=columns)

    df1_final['Bits']=x_bits
    df1_final['Time']=cum_times
    df1_final['CumBits']=cum_bits

    # Adversary
    adv_file=folder_adv+f"{i}"+".csv"
    df2 = pd.read_csv(adv_file, skiprows=1)
    df2['Delta Timestamp']=df2[' Time Stamp'].diff()/1000
    df2['Delta Timestamp']=df2['Delta Timestamp'].fillna(0)
    df2['CumBits']=df2[' Extra Bits'].cumsum()
    df2['CumTime']=df2['Delta Timestamp'].cumsum()

    rows, columns = df2.shape
    y_bits=[]
    for i in range(0, rows):
        y_bits.append(i/math.log(2))

    columns=['Bits', 'Time', 'CumBits']
    df2_final = pd.DataFrame(columns=columns)

    df2_final['Bits']=y_bits
    df2_final['Time']=cum_times
    df2_final['CumBits']=cum_bits

    results = []

    for bits in df1_final['Bits']:
        row_dfh = df1_final[df1_final['Bits'] == bits]
        row_dfa = df2_final[df2_final['Bits'] == bits]

        winner = 1 if row_dfh['Time'].iloc[0] < row_dfa['Time'].iloc[0] else 0
        results.append([bits, winner])

    winning_df = pd.DataFrame(results, columns=['Bits', 'Winner'])

    return winning_df
```

```python
compute_one_round(2)
```

```python
TOTAL_ROUNDS=100
all_dfs=[]
for i in range(1, TOTAL_ROUNDS):
    all_dfs.append(compute_one_round(i))
```

```python
all_dfs[3]
```

```python
m_bits=[]
for i in range(0, TOTAL_ROUNDS):
    m_bits.append(i/math.log(2))

total_winner=[]
for bits in m_bits:
    result=0
    for df in all_dfs:
        for value in df[df['Bits']==bits]['Winner']:
            result+=value
    total_winner.append(result)

columns=['Bits', 'Total Wins']
m_final = pd.DataFrame(columns=columns)
m_final['Bits']=m_bits
m_final['Total Wins']=total_winner
```

```python
m_final
```

```python
m_final.to_csv("final.csv", index=False)
```
